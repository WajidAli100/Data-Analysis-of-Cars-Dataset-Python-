# Data Analysis of Cars Dataset

```python
import pandas as pd
data=pd.read_csv('cars.csv')
data.head(2)
```

```python
data.describe()
```

* Summary statistics: Get an overview of the data

```python
data.head(2)
```

* Data Cleaning
* Handling missing values: Check for and handle missing data

```python
data.isnull().sum() # check for null values
```

```python
data['Cylinders'].fillna(method='ffill', inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Length'].fillna(data['Length'].mean(),inplace=True)   
```

```python
data.isnull().sum()
```

* Find different/unique car makers in the dataset

```python
data['Make'].value_counts()
```

* Data Transformation
* Creating new columns: For example, calculating the power-to-weight ratio

```python
data['power-to-weight']=data['Horsepower']/data['Weight']
data['power-to-weight']
```

* Filtering and Sorting Data
* Filtering rows: Get cars with specific criteria

```python
luxury_cars=data[data['Horsepower']>300]
luxury_cars
```

```python
luxury_cars=luxury_cars[['Make','Model','Horsepower']]  
luxury_cars
```

```python
data.sort_values(by='Horsepower', ascending=False)
```

* Find all the records where origin is Asia or America

```python
data[data['Origin'].isin(['Asia','America'])]
```

* Group By Operations

```python
data.groupby('Make')['Cylinders'].mean()
```

```python
data.groupby('Type').agg({'Horsepower':'max','MPG_City':'mean'})
```

* Find all the records where wait is above 4000

```python
data[data['Weight']>4000]
```

* Visualization
* Using pandas built-in plotting: Simple visualizations.

```python
import matplotlib.pyplot as plt
data['Horsepower'].hist()
plt.xlabel('Horsepower')
plt.ylabel('Frequency')
plt.title('Horsepower Distribution')
plt.show()
```

* Correlation and Covariance

```python
import numpy as np
data = data.select_dtypes(include=[np.number])
data.corr()

```

* Pivot Tables

```python
pd.pivot_table(data, values='Horsepower', index='Make', columns='Type', aggfunc=np.mean)
```
